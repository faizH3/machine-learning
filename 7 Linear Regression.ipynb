# Machine Learning - Linear Regression

### 1 Regression

Istilah regresi digunakan ketika mencoba menemukan hubungan antar variabel.

Dalam machine learning dan pemodelan statistik digunakan untuk memprediksi hasil peristiwa di masa mendatang.

### 2 Linear Regression

Regresi linear(Linear regression) menggunakan hubungan antar titik data untuk menggambar garis lurus. Baris ini dapat digunakan untuk memprediksi nilai dimasa mendatang.

```python
import matplotlib.pyplot as plt
#####################################
"""
sumbu x mewakili usia
sumbu y mewakili kecepetan
daftar usia dan kecepatan 13 mobil saat melewati pintu tol.
"""
#####################################


x = [5,7,8,7,2,17,2,9,4,11,12,9,6]  # age
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]  # speed

plt.scatter(x, y)
plt.show()
```

```python
import matplotlib.pyplot as plt
from scipy import stats

x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

slope, intercept, r, p, std_err = stats.linregress(x, y)

def myfunc(x):
  return slope * x + intercept

mymodel = list(map(myfunc, x))

plt.scatter(x, y)
plt.plot(x, mymodel)
plt.show()
```

Jika antara nilai sumbu x dan y, maka regresi linear tidak dapat digunakan untuk memprediksi.

`r` nilainya berkisar antara -1 dan 1, dimana 0 berarti tidak ada hubungan dan 1 & -1 berarti 100%.



### 4 Predict Future Values

Prediksi nilai masadepan

```python
# memprediksi kecepatan mobil usia 10 tahun
from scipy import stats

x = [5,7,8,7,2,17,2,9,4,11,12,9,6]  # usia
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]  # kecepatan

slope, intercept, r, p, std_err = stats.linregress(x, y)

def myfunc(x):
  return slope * x + intercept

speed = myfunc(10)

print(speed)
```

```python
import matplotlib.pyplot as plt
from scipy import stats

x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

slope, intercept, r, p, std_err = stats.linregress(x, y)

def myfunc(x):
  return slope * x + intercept

mymodel = list(map(myfunc, x))

plt.scatter(x, y)
plt.plot(x, mymodel)

# usia, prediksi
plt.scatter(10, 85.59308314937454)

plt.show()
```
